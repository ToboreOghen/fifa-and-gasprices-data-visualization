```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa.overall, bins = bins)


plt.xticks(bins)

plt.xlabel('Player Skill Level')
plt.ylabel('Number Of Players')
plt.title('Distribution Of Players Skill In FIFA(2022)', fontdict = {'fontweight': 'bold','fontsize':15})

plt.show()
```

```python

#This is What happens when you use none existing columns.

left = fifa.loc[ fifa['preferred_foot'] == 'Left'].shape[0]
right = fifa.loc[ fifa['preferred_foot'] == 'Right'].shape[0]

labels= ['Left', 'Right']
colors = ['#abcdef', '#aabbcc']

plt.pie([left,right], labels = labels, colors = colors, autopct ='%.2f %%')

plt.title('Foot Preference Of FIFA players')
plt.show()
```

```python
light = fifa.loc[fifa.weight_kg < 56].shape[0]
light_medium = fifa.loc[(fifa.weight_kg >= 56) & (fifa.weight_kg < 68)].shape[0]
medium = fifa[(fifa.weight_kg >= 68) & (fifa.weight_kg < 79)].shape[0]
medium_heavy = fifa[(fifa.weight_kg >=79 ) & (fifa.weight_kg < 90)].shape[0]
heavy = fifa[(fifa.weight_kg >= 90)].shape[0]

weights = [light, light_medium, medium, medium_heavy, heavy]
labels= ['under 56', '56-68', '68-79','79-90', 'above 90']
explode = (.4,.1,0,0,.4)

plt.style.use('ggplot')
plt.pie( weights, labels= labels, autopct = '%.2f %%', pctdistance = 0.8, explode= explode)
plt.title ('Weight Distribution Of FIFA Players(In Kg)')

plt.show()
```

```python
plt.figure(figsize=(5,8) )


barcelona = fifa.loc[fifa.club_name == 'FC Barcelona']['overall']
madrid = fifa.loc[fifa.club_name == 'Real Madrid CF']['overall']
chelsea = fifa.loc[fifa.club_name == 'Chelsea']['overall']
man_u = fifa.loc[fifa.club_name == 'Manchester United']['overall']

club = [barcelona, madrid, chelsea, man_u]
labels = ['FC Barcelona', 'Real Madrid', 'Chelsea','Man. United']


plt.style.use('default')

boxes = plt.boxplot(club, tick_labels = labels, patch_artist = True, medianprops={'linewidth' : 2})

for box in boxes ['boxes']:
    box.set(color = '#4286f4', linewidth= 2.5)

    box.set(facecolor = '#e0e0e0')


plt.title ('FIFA Football Club Comparison')
plt.ylabel('FIFA Overall Rating')
plt.xlabel('FIFA Club')
plt.show()
```
